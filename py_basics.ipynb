```python
import numpy as np
from matplotlib import pyplot as plt

print("pi = ", np.pi)
```

```python
def seno_verdadero(t):
    return np.sin(t)

def seno_aprox(t):
    return t - (1 / 6) * t ** 3

cant_muestras = 50
t = np.linspace(-np.pi / 2, np.pi / 2, cant_muestras)

x_verdadero = seno_verdadero(t)
x_aprox = seno_aprox(t)

plt.figure()
plt.plot(t, x_aprox, 'k.', lw = 3, label = 'seno_aprox(pi*t)')
plt.plot(t, x_verdadero, 'k.', lw = 3, label = 'seno_verdadero(pi*t)')

plt.xlabel('t/pi')
plt.ylabel('x(t)')

plt.title('aproximacion seno')

plt.legend(loc='best')
plt.grid(True)
plt.show()

plt.savefig("sen_aprox.jpg")
```

```python
diferencia_aprox = np.abs(x_aprox - x_verdadero)

plt.figure()
plt.plot(t, diferencia_aprox)
plt.xlabel('t/pi')
plt.ylabel('dif(t)')
plt.title("diferencia en aprox")
 
```
