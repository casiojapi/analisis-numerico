```python
# encontrar la solucion de x^3 + 4* x^2 - 10 = 0 entre x [1, 2]
## deberia ser una cosa similar a busqueda binaria.

import numpy as np


def f(x):
    return x ** 3 + 4 * x ** 2 - 10


def biseccion(f, a, b, TOL, N0):
    if f(a) * f(b) >= 0:
        print("no puede existir una raiz en el intervalo dado")
        return None
    
    p_anterior = a
    
    i = 1
    while (i < N0):
        p = (a + b) / 2
        if np.abs(f(p) - f(p_anterior)) < TOL:
            return p, i
        
        if f(a) * f(p) > 0:
            a = p
        else:
            b = p
        i = i + 1 
        p_anterior = p
    
    print("no converge")
    return None

#test
tolerancia = 1e-4
n_max = 25
a = 1 
b = 2
raiz, n_iteraciones = biseccion(f, a, b, tolerancia, n_max)

print("la raiz encontrada es: ", raiz, " usando N = ", n_iteraciones, " iteraciones")


def biseccion_historial(f, a, b, TOL, N0):
    if f(a) * f(b) >= 0:
        print("no puede existir una raiz en el intervalo dado")
        return None
    
    p_anterior = a
    historial = []
    
    i = 1
    while (i < N0):
        p = (a + b) / 2
        
        historial.append((i, p))
        
        if np.abs(f(p) - f(p_anterior)) < TOL:
            return p, i, historial
        
        if f(a) * f(p) > 0:
            a = p
        else:
            b = p
        i = i + 1 
        p_anterior = p
    
    print("no converge")
    return None

#test
tolerancia = 1e-4
n_max = 25
a = 1 
b = 2
raiz, n_iteraciones, historial = biseccion_historial(f, a, b, tolerancia, n_max)

print("----------------------------------------------------------")
print("iteracion", "\t", "raiz", "\t", "tolerancia: ", tolerancia)
print("----------------------------------------------------------")
for i in historial: 
    print(i[0], "\t\t\t", i[1])
```
